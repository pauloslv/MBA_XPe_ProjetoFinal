```python
dados = pd.read_csv('/content/Stock_Screener.csv', sep=',', encoding='UTF-8', skiprows=1)
```

```python
#Converter dados da coluna 'Volume Financ.(R$)' para float
```

```python
dados
```

```python
dados.loc['Volume Financ.(R$)'] = dados['Volume Financ.(R$)'].apply(lambda x: str(x).replace(".",""))
```

```python
dados ['Volume Financ.(R$)'].astype('float64')
```

```python
#Filtrando empresas com negociação maior que 1.000.000 por dia
```

```python
display(dados.loc[(dados['Volume Financ.(R$)'] >=1000000)])
```

```python
#Excluir empresas com negociação menor que 1.000.000
```

```python
Liquidez = (dados.loc[(dados['Volume Financ.(R$)'] < 1000000)])
```

```python
dados = dados.drop(Liquidez.index)
```

```python
dados['Margem EBIT'] = dados['Margem EBIT'].apply(lambda x: str(x).replace(".",""))
```

```python
dados['Margem EBIT'] = dados['Margem EBIT'].apply(lambda x: str(x).replace("%",""))
```

```python
dados['Margem EBIT'] = dados['Margem EBIT'].astype('float64')
```

```python
#Excluir empresas com Margem EBIT negativa
```

```python
MargemEBITNeg = dados.loc[(dados['Margem EBIT'] < 0)]
```

```python
dados = dados.drop(MargemEBITNeg.index)
```

```python
#Ordenar as empresas pelo EV/EBIT
```

```python
dados_ordenadoEV = dados.sort_values(by='EV/EBIT', ascending=True)
```

```python
dados_ordenadoEV.head(20) 
```

```python
#Exclusão das empresas em recuperação judicial
```

```python
rec_jud = ['ATMP3', 'BDLL3', 'BDLL4', 'ETER3', 'FRTA3', 'HOOT4', 'IGBR3', 'INEP3', 'INEP4', 'JFEN3', 'LUPA3', 'OIBR3', 'OIBR4', 'RNEW3', 'RNEW4', 'RNEW11', 'RPMG3', 'RSID3', 'SLED3', 'SLED4', 'TEKA3', 'TEKA4']
```

```python
excluir = dados_ordenadoEV.loc[dados['Ação'].isin(rec_jud)]
```

```python
dados_final = dados_ordenadoEV.drop(excluir.index)
```

```python
dados_final.head(20)
```

```python
#Excluir ações da mesma empresa
```

```python
dados_final.drop_duplicates(subset = 'Empresa', keep = 'first', inplace= True)
```
